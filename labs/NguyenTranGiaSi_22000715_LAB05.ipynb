```python
# prompt: Bài 04: Mô phỏng tung một con xúc sắc cân đối đồng chất 5000 lần. Dựa vào giá trị mô phỏng, bạn hãy tìm các giá trị xác suất dưới đây:
# a. Xác suất để số chấm xuất hiện là 4
# b. Xác suất để số chấm xuất hiện lớn hơn hoặc bằng 4
# c. Giả sử biết số chấm xuất hiện lớn hơn hoặc bằng 4. Hãy tìm xác suất để mặt 6 chấm xuất hiện.

# a. Xác suất để số chấm xuất hiện là 4
import numpy as np

# Mô phỏng tung xúc xắc 5000 lần
num_trials = 5000
rolls = np.random.randint(1, 7, size=num_trials)

count_4 = sum(rolls == 4)
print(f"Số lần mặt 4 chấm xuất hiện: {count_4}")
xac_xuat_4 = count_4 / num_trials
print(f"Xác suất để số chấm xuất hiện là 4: {xac_xuat_4}")

# b. Xác suất để số chấm xuất hiện lớn hơn hoặc bằng 4
count_greater_than_or_equal_4 = sum(rolls >= 4)
print(f"Số lần mặt 4 chấm xuất hiện hoặc lớn hơn: {count_greater_than_or_equal_4}")
xac_xuat_greater_than_or_equal_4 = count_greater_than_or_equal_4 / num_trials
print(f"Xác suất để số chấm xuất hiện lớn hơn hoặc bằng 4: {xac_xuat_greater_than_or_equal_4}")

# c. Giả sử biết số chấm xuất hiện lớn hơn hoặc bằng 4. Hãy tìm xác suất để mặt 6 chấm xuất hiện.
count_6 = sum(rolls == 6)
print(f"Số lần mặt 6 chấm xuất hiện: {count_6}")
xac_xuat_6 = count_6 / count_greater_than_or_equal_4
print(f"Xác suất để mặt 6 chấm xuất hiện: {xac_xuat_6}")

```

Giả sử X là biến ngẫu nhiên rời rạc

a. Hãy cho biết ý nghĩa câu lệnh sau: `binom.pmf(4, 10, 0.6)`

b. Giả sử X có phân phối nhị thức `X ~ B(10, 0.6)`. Hãy tính các giá trị sau:  
$P(X \leq 5), P(X \le 5), P(X \ge 4), P(X = 5) $

c. Hãy cho biết ý nghĩa câu lệnh sau: `poisson.pmf(4, 3)`

d. Giả sử X có phân phối Poisson `X ~ P(3)`. Hãy tính các giá trị sau:  
$P(X \leq 5), P(X \le 5), P(X \ge 4), P(X = 5) $

```python
# a. Ý nghĩa câu lệnh binom.pmf(4, 10, 0.6)
# Câu lệnh này tính xác suất khối lượng xác suất của một biến ngẫu nhiên X tuân theo phân phối nhị thức với tham số n = 10), p = 0.6 tại giá trị X = 4.
# Nói cách khác, nó tính xác suất có đúng 4 lần thành công trong 10 lần thử, với xác suất thành công ở mỗi lần thử là 0.6.

# b. Tính các giá trị xác suất cho phân phối nhị thức X ~ B(10, 0.6)
n = 10
p = 0.6

# P(X ≤ 5)
P_X_le_5 = binom.cdf(5, n, p)
print(f"P(X ≤ 5) = {P_X_le_5}")

# P(X ≥ 4)
P_X_ge_4 = 1 - binom.cdf(3, n, p)
print(f"P(X ≥ 4) = {P_X_ge_4}")

# P(X = 5)
P_X_eq_5 = binom.pmf(5, n, p)
print(f"P(X = 5) = {P_X_eq_5}")

# c. Ý nghĩa câu lệnh poisson.pmf(4, 3)
# Câu lệnh này tính xác suất khối lượng xác suất của một biến ngẫu nhiên X tuân theo phân phối Poisson với tham số λ = 3 tại giá trị X = 4.
# Nó cho biết xác suất để có đúng 4 sự kiện xảy ra trong một khoảng thời gian hoặc không gian xác định, khi biết trung bình có 3 sự kiện xảy ra trong
#  khoảng thời gian hoặc không gian đó.

# d. Tính các giá trị xác suất cho phân phối Poisson X ~ P(3)
mu = 3

# P(X ≤ 5)
P_X_le_5 = poisson.cdf(5, mu)
print(f"P(X ≤ 5) = {P_X_le_5}")

# P(X ≥ 4)
P_X_ge_4 = 1 - poisson.cdf(3, mu)
print(f"P(X ≥ 4) = {P_X_ge_4}")

# P(X = 5)
P_X_eq_5 = poisson.pmf(5, mu)
print(f"P(X = 5) = {P_X_eq_5}")
```

- `norm.ppf(0.975)`

- `norm.ppf(0.975, 1, 1)`

d. Giả sử biến ngẫu nhiên X có phân phối chuẩn với kỳ vọng là 24 và phương sai là 16. Hãy tính các giá trị sau:

- $P(X \leq 20)$
- $P(X \ge 29.5)$
- $P(X = 23.8)$

```python
# a.
# norm.cdf(2): Xác suất cho biến chuẩn chuẩn tắc nhỏ hơn hoặc bằng 2.

# norm.cdf(2, 1, 1): Xác suất cho biến chuẩn với trung bình = 1, độ lệch chuẩn = 1 nhỏ hơn hoặc bằng 2.

# norm.cdf(2, 1, 2): Xác suất cho biến chuẩn với trung bình = 1, độ lệch chuẩn = 2 nhỏ hơn hoặc bằng 2.

# norm.ppf(norm.cdf(2)): Trả về giá trị 2, vì đây là hàm nghịch đảo của CDF.

print(norm.ppf(norm.cdf(2)))

# c. Tính các giá trị và vẽ hình minh họa
# norm.ppf(0.975)
ppf_val1 = norm.ppf(0.975)
print(f"norm.ppf(0.975) = {ppf_val1}")

# norm.ppf(0.975, 1, 1)
ppf_val2 = norm.ppf(0.975, loc=1, scale=1)
print(f"norm.ppf(0.975, 1, 1) = {ppf_val2}")

# norm.ppf(0.975, 1, 2)
ppf_val3 = norm.ppf(0.975, loc=1, scale=2)
print(f"norm.ppf(0.975, 1, 2) = {ppf_val3}")

# Vẽ hình minh họa
x = np.linspace(-4, 4, 1000)
plt.plot(x, norm.pdf(x), label="N(0,1)")
plt.plot(x, norm.pdf(x, loc=1, scale=1), label="N(1,1)")
plt.plot(x, norm.pdf(x, loc=1, scale=2), label="N(1,2)")

plt.axvline(ppf_val1, color='red', linestyle='dashed', linewidth=1)
plt.axvline(ppf_val2, color='green', linestyle='dashed', linewidth=1)
plt.axvline(ppf_val3, color='blue', linestyle='dashed', linewidth=1)

plt.legend()
plt.title('Minh họa các giá trị ppf')
plt.show()

# d. Tính các giá trị xác suất
mu = 24
sigma = 4  # Độ lệch chuẩn (căn bậc hai của phương sai)

# P(X ≤ 20)
prob1 = norm.cdf(20, loc=mu, scale=sigma)
print(f"P(X ≤ 20) = {prob1}")

# P(X ≥ 29.5)
prob2 = 1 - norm.cdf(29.5, loc=mu, scale=sigma)
print(f"P(X ≥ 29.5) = {prob2}")

# P(X = 23.8)
# Xác suất của một biến ngẫu nhiên liên tục tại một điểm cụ thể luôn bằng 0.
prob3 = 0
print(f"P(X = 23.8) = {prob3}")
```

**Bài 8**

Một công ty sản xuất làm ra các chi tiết máy với độ dài các chi tiết có phân phối chuẩn với kỳ vọng là 3000 mm và độ lệch chuẩn là 3 mm. Yêu cầu cho các chi tiết máy này là phải có chiều dài từ 2993 mm đến 3007 mm. Tính tỷ lệ sản phẩm lỗi của công ty

```python
mu = 3000
sigma = 3

# Giới hạn chấp nhận được
thap = 2993
cao = 3007

# Tính xác suất sản phẩm nằm trong giới hạn chấp nhận được
ti_le_sai_duoc = norm.cdf(cao, mu, sigma) - norm.cdf(thap, mu, sigma)

# Tỷ lệ sản phẩm lỗi
ti_le_loi = 1 - ti_le_sai_duoc

print(f"Tỷ lệ sản phẩm lỗi của công ty là: {ti_le_loi}")
```

**Bài 9**

Giả sử điểm thi TOEIC là một biến ngẫu nhiên có phân phối chuẩn với kỳ vọng là 500, độ lệch chuẩn là 8. Bạn hãy:

a. Mô phỏng bằng cách phát sinh ngẫu nhiên điểm thi TOIEC của một nhóm gồm 50 người.

b. Tính điểm trung bình và độ lệch chuẩn về điểm của nhóm.

c. Tìm miền giá trị, và miền phân vị (IQR) của nhóm

d. Cho biết tỷ lệ đạt trên 450 điểm của nhóm.

e. Vẽ đồ thị histogram của nhóm và so sánh với đồ thị hàm mật độ của phân phối chuẩn với kỳ vọng là 500 và độ lệch chuẩn là 8.

f. Thay vì mô phỏng 50 người, hãy mô phỏng nhóm với 100 người thi. Bạn có nhận xét gì?

```python
# a. Mô phỏng bằng cách phát sinh ngẫu nhiên điểm thi TOEIC của một nhóm gồm 50 người.
mu = 500
sample = 50
std = 8

toeic_50 = np.random.normal(loc=mu, scale=std, size=sample)
print("Điểm thi TOEIC của 50 người: " + str(toeic_50))

# b. Tính điểm trung bình và độ lệch chuẩn về điểm của nhóm.
print("Điểm trung bình: " + str(np.mean(toeic_50)))
print("Độ lệch chuẩn: " + str(np.std(toeic_50)))

# c. Tìm miền giá trị, và miền phân vị (IQR) của nhóm
mien_gia_tri = np.min(toeic_50), np.max(toeic_50)
q1 = np.quantile(toeic_50, 0.25)
q3 = np.quantile(toeic_50, 0.75)
iqr = q3 - q1

print(f"Miền giá trị của nhóm: {mien_gia_tri}")
print(f"IQR của nhóm: {iqr}")

# d. Cho biết tỷ lệ đạt trên 450 điểm của nhóm.
toeic_450 = 0
for score in toeic_50:
    if score > 450:
        toeic_450 += 1

ti_le_450 = toeic_450 / sample
print(f"Tỷ lệ đạt trên 450 điểm: {ti_le_450}")

# e. Vẽ đồ thị histogram của nhóm và so sánh với đồ thị hàm mật độ của phân phối chuẩn với kỳ vọng là 500 và độ lệch chuẩn là 8
plt.hist(toeic_50, color="blue", bins=50, label="Histogram của nhóm")

# f. Thay vì mô phỏng 50 người, hãy mô phỏng nhóm với 100 người thi. Bạn có nhận xét gì?
# => Khi mô phỏng nhóm 100 người, ta sẽ thấy rằng điểm trung bình và độ lệch chuẩn có xu hướng gần với kỳ vọng hơn,
#    vì kích thước mẫu lớn hơn, giúp ước lượng chính xác hơn các tham số của phân phối chuẩn.
```

Đối với mỗi N từ 10; 40; 90; 160; 250; 490; 640; 810; 1000, hãy ước tính giá trị của X bằng cách mô phỏng số lần tung đó.

Bạn nên chạy mỗi mô phỏng 100 lần và sử dụng tập hợp các ước tính để đánh giá phương sai của ước tính x của bạn. Vẽ biểu đồ phương sai này so với N, bạn có nhận xét gì không?

```python
N_gia_tri = [10, 40, 90, 160, 250, 490, 640, 810, 1000]
phuong_sai = []

so_lan = 100

for N in N_gia_tri:
    uoc_tinh = [np.mean(np.random.randint(0, 2, size=N)) for _ in range(so_lan)]
    phuong_sai_n = np.var(uoc_tinh)
    phuong_sai.append(phuong_sai_n)

plt.plot(N_gia_tri, phuong_sai, marker='o')
plt.xlabel('Số lần tung (N)')
plt.ylabel('Phương sai của ước tính X')
plt.title('Phương sai của ước tính xác suất tung được mặt Head theo số lần tung')
plt.show()

# Nhận xét: Khi số lần tung N tăng lên, phương sai của ước tính X giảm dần.
# Điều này thể hiện rằng với số lần tung lớn hơn, ta sẽ có được ước tính chính xác hơn về xác suất tung được mặt Head.
```
